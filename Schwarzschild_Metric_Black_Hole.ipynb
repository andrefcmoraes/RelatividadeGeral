```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
from google.colab import files
```

```python
Rs = 5
x0 = 40
radious = 1.5*Rs
def function(x,tau):
  return [-(Rs/(Rs*x[3]-x[3]**2))*x[0]*x[2],
          x[0],
          -0.5*(Rs/(x[3]**2)-(Rs**2)/(x[3]**3))*x[0]**2 -0.5*(-Rs/(x[3]-Rs)**2 + Rs**2/(x[3]*(x[3]-Rs)**2))*x[2]**2-(Rs-x[3])*x[4]**2,
          x[2],
          -(2/x[3])*x[2]*x[4],
          x[4]]
tau = np.linspace(0,150,10**6)
################################################################################
Z = []
F = []
NX = []
NY = []
PN = int(20)
def f(b):
  ci = [1/np.sqrt(1-Rs/np.sqrt(x0**2+b**2)), 0, -x0/np.sqrt(x0**2+b**2), np.sqrt(x0**2+b**2), 0.004, np.arccos(x0/np.sqrt(x0**2+b**2))]
  X = odeint(function, ci, tau)
  return Z.append(X)

for i in range(PN):
  f(1.7*i+5)
  NX.append(Z[i][:,3]*np.cos(Z[i][:,5]))
  NY.append(Z[i][:,3]*np.sin(Z[i][:,5]))

figure, axes = plt.subplots()
for n in range(PN):
  plt.plot(NX[n],NY[n], color = 'black')
draw_circle = plt.Circle((0, 0), Rs, color='black')
axes.set_aspect(1)
draw_circle2 = plt.Circle((0,0), 1.5*Rs,fill=False,ec="red")
axes.set_aspect(1)
axes.add_artist(draw_circle2)
axes.add_artist(draw_circle)
plt.ylim((-40,40))
plt.xlim((-40,40))
plt.title('Buraco negro Rs = 5')
plt.savefig('grafico.pdf')
files.download("grafico.pdf") 
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
%matplotlib inline
from google.colab import files
```

```python
Rs = 5
x0 = 65
radious = 1.5*Rs
Z = []
F = []
NX = []
NY = []
PN = int(20)
def f(b):
  r0 = np.sqrt(x0**2+b**2)
  phi0 = np.arccos(x0/np.sqrt(x0**2+b**2))
  p0 = x0/r0
  L = r0*np.sin(phi0)
  def function(x, l):
    r = x[0]
    p = x[1]
    phi = x[2]
    return [p, L**2*(r-1.5*Rs)/r**4, L/r**2]
  l = np.linspace(0,150,10**6)
  ci = [r0, -x0/np.sqrt(x0**2+b**2), phi0]
  X = odeint(function, ci, l)
  return Z.append(X)

for i in range(PN):
  f(i+10)
  NX.append(Z[i][:,0]*np.cos(Z[i][:,2]))
  NY.append(Z[i][:,0]*np.sin(Z[i][:,2]))

figure, axes = plt.subplots()
for n in range(PN):
  plt.plot(NX[n],NY[n], color = 'black')
draw_circle = plt.Circle((0, 0), Rs, color='black')
axes.set_aspect(1)
draw_circle2 = plt.Circle((0,0), 1.5*Rs,fill=False,ec="red")
axes.set_aspect(1)
axes.add_artist(draw_circle2)
axes.add_artist(draw_circle)
plt.ylim((-50,50))
plt.xlim((-50,50))
plt.title('Buraco negro Rs = 4')
plt.show()
```
