```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
from google.colab import files
%matplotlib inline
```

```python
Rs = 1
def function(x,tau):
  return [-(Rs/(Rs*x[3]-x[3]**2))*x[0]*x[2],
          x[0],
          -0.5*(Rs/(x[3]**2)-(Rs**2)/(x[3]**3))*x[0]**2 -0.5*(-Rs/(x[3]-Rs)**2 + Rs**2/(x[3]*(x[3]-Rs)**2))*x[2]**2-(Rs-x[3])*x[4]**2,
          x[2],
          -(2/x[3])*x[2]*x[4],
          x[4]]
tau = np.linspace(0,200,10**6)
radious = 5
ci = [1/np.sqrt(1-Rs/radious), 0, 0, radious, 0.09, 0]
X = odeint(function, ci, tau)
x = X[:,3]*np.cos(X[:,5])
y = X[:,3]*np.sin(X[:,5])
figure, axes = plt.subplots()
draw_circle = plt.Circle((0, 0), Rs, color='black')
plt.plot(x,y)
axes.set_aspect(1)
draw_circle2 = plt.Circle((0,0), 1.5*Rs,fill=False,ec="red")
axes.set_aspect(1)
axes.add_artist(draw_circle2)
axes.add_artist(draw_circle)
plt.title('Buraco negro Rs = 1')
plt.show()
```
